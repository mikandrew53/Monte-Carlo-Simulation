# Example Script for Monte Carlo and Importance 
The purpose of this notebook is to suplement my tutorial on Monte Carlo and importance sampling. Feel free to use and adapt any of these functions for your own numeric simulations.

```python
import numpy as np
import math
import random
from matplotlib import pyplot as plt
from IPython.display import clear_output

```

## Baseline Functions

```python
def get_rand_number(min_value, max_value):
    range = max_value - min_value
    choice = random.uniform(0,1)
    return min_value + range*choice
```

```python
def getGame():
    return [
        {
            "player1": [],
            "player2": [],
            "heads": 0,
            "tails": 0,
        },
        {
            "player1": [],
            "player2": [],
            "heads": 0,
            "tails": 0,
            
        },
        {
            "player1": [],
            "player2": [],
            "heads": 0,
            "tails": 0,
        },
        {
            "player1": [],
            "player2": [],
            "heads": 0,
            "tails": 0,
        },
    ]
```

```python
def dart1():
    lower_bound = 1
    upper_bound = 101
    games = getGame()
    for i in range(1,5):
        num_samples = ((10)**i)
        for j in range(0, num_samples):
            coin = get_rand_number(1, 100)
            if coin >= 1 and coin <= 50:
                player = "player1"
                games[i-1]["heads"] += 1
            elif coin >= 51 and coin <= 100:
                games[i-1]["tails"] += 1
                player = "player2"
                
            x = get_rand_number(lower_bound, upper_bound)    
            if x <= 4:
                games[i-1][player].append(-1) 
            elif x >= 5 and x <= 15:
                games[i-1][player].append(50)
            elif x >= 16 and x <= 47:
                games[i-1][player].append(10)
            elif x >= 48 and x <= 101:
                games[i-1][player].append(1)

   
    for i in range(4):
        scores = ((10)**(i+1))
        plt.hist(games[i]["player1"],edgecolor="black")
        plt.title('Part 1 Scores ' + str(scores))
        plt.xlabel('Score')
        plt.ylabel('Frequency')
        plt.tight_layout()
        plt.show()

        


```

```python
def dart2():
    lower_bound = 1
    upper_bound = 101
    games = getGame()
    for i in range(1,5):
        num_samples = ((10)**i)
        for j in range(0, num_samples):
            games[i-1]["player1"].append(0)
            games[i-1]["player2"].append(0)
            for k in range(4):
                coin = get_rand_number(1, 100)
                if coin >= 1 and coin <= 50:
                    player = "player1"
                    games[i-1]["heads"] += 1
                elif coin >= 51 and coin <= 100:
                    games[i-1]["tails"] += 1
                    player = "player2"

                x = get_rand_number(lower_bound, upper_bound)
                
                if x <= 4:
                    games[i-1][player][j] -= 1
                elif x >= 5 and x <= 15:
                    games[i-1][player][j] += 50
                elif x >= 16 and x <= 47:
                    games[i-1][player][j] += 10
                elif x >= 48 and x <= 101:
                    games[i-1][player][j] += 1

   
    for i in range(4):
        scores = ((10)**(i+1))
        plt.hist(games[i]["player1"], edgecolor="black")

        plt.title('Part 2 Scores ' + str(scores))
        plt.xlabel('Score')
        plt.ylabel('Frequency')
        plt.tight_layout()
        plt.show()
```

```python
dart1()
dart2()
```
